## Описание данных

Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.

Стадии:
1. Флотация
Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).
2. Очистка
Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

Технологический процесс:
- Rougher feed — исходное сырье
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
- Xanthate — ксантогенат (промотер, или активатор флотации);
- Sulphate — сульфат (на данном производстве сульфид натрия);
- Depressant — депрессант (силикат натрия).
- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота

Параметры этапов:
- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи

Наименование признаков:
[этап].[тип_параметра].[название_параметра]
Возможные значения для блока [этап]:
- rougher — флотация
- primary_cleaner — первичная очистка
- secondary_cleaner — вторичная очистка
- final — финальные характеристики

Возможные значения для блока [тип_параметра]:
- input — параметры сырья
- output — параметры продукта
- state — параметры, характеризующие текущее состояние этапа
- calculation — расчётные характеристики

### Проверка правильности расчет эфективности обогощения

Попробуем вычислить эфективность на обучающей выборке - df_train для признака rougher.output.recovery.  
Сравним Абсолютную среднюю ошибку между расчётами и значением признака.  
Расчет производится по формуле:

Где:
 - C — доля золота в концентрате после флотации/очистки;
 - F — доля золота в сырье/концентрате до флотации/очистки;
 - T — доля золота в отвальных хвостах после флотации/очистки.

Напишем функцию проверки по формуле:

```python
def recovery_check(c, f, t):
    recovery = ((c * (f - t)) / (f * (c - t))) * 100
    return recovery
```

```python
recovery = recovery_check(
    df_train['rougher.output.concentrate_au'],
    df_train['rougher.input.feed_au'],
    df_train['rougher.output.tail_au'])

mean_absolute_error(df_train['rougher.output.recovery'], recovery)
```

#### Вывод

Абсолютная средняя ошибка получилась очень маленькая 9.73e-15, это значит, что данные рассчитаны верно.

## Анализ данных

### Посмотрим, как меняется концентрация металлов (Au, Ag, Pb) на различных этапах очистки.

```python
def metall_concentration(rougher_input, rougher_output, primary_cleaner, final, xlim, name, leg_loc, nbins):
    sns.set_theme(style='whitegrid', palette='tab10', color_codes=True)
    plt.figure(figsize=(15,5))
    ax = sns.distplot(rougher_input, bins = nbins)
    ax = sns.distplot(rougher_output, bins = nbins)
    ax = sns.distplot(primary_cleaner, bins = nbins)
    ax = sns.distplot(final, bins = nbins)
    ax.set_xlim(xlim)
    plt.title(name)
    plt.legend(
        title='этапы',
        loc=leg_loc,
        labels=['до флотации', 'после флотации', 'после первичной очистки', 'итоговая'])
    plt.xlabel("концентрация (%)")
    plt.ylabel("количество")
    plt.show()
    print('средняя концентрация до флотации:{: 0.2f}%'.format(rougher_input.mean()))
    print('средняя концентрация после флотации:{: 0.2f}%'.format(rougher_output.mean()))
    print('средняя концентрация после первичной очистки:{: 0.2f}%'.format(primary_cleaner.mean()))
    print('средняя итоговая концентрация:{: 0.2f}%'.format(final.mean()))
```

### Напишем функцию для вычисления итогового sMAPE

Симметричная средняя абсолютная процентная ошибка (sMAPE) используется для измерения прогностической точности моделей.

В Python нет встроенной функции для вычисления sMAPE, но мы можем создать для этого простую функцию, где:
- target - фактическое значение данных
- predict - прогнозируемое значение данных

```python
def sMAPE (target, predictions):
    sMAPE = (sum(abs(target - predictions) / ((abs(target) + abs(predictions)) / 2)) / len(target)) * 100
    return sMAPE
```

Итоговая метрика складывается из 2-х величин sMAPE:

Напишем функцию и для итоговой величины:

```python
def final_sMAPE(sMAPE_rougher, sMAPE_final):
    f_sMAPE = -1 * (pd.Series( 0.25 * sMAPE_rougher + 0.75 * sMAPE_final).mean())
    return f_sMAPE
```
