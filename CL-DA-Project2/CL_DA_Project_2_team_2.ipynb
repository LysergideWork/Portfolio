```python
import pandas as pd
import plotly.express as px
```

```python
df1 = pd.read_csv('/content/Datafiniti_Fast_Food_Restaurants.csv')
df2 = pd.read_csv('/content/Datafiniti_Fast_Food_Restaurants_May19.csv')
df3 = pd.read_csv('/content/FastFoodRestaurants.csv')
```

```python
df_combined = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
df_combined
```

```python
df_droped = df_combined.drop(columns=['id','keys','sourceURLs','websites','primaryCategories'])
df_droped
```

```python
df_droped.info()
```

```python
duplicates = df_droped[df_droped.duplicated(subset=['latitude', 'longitude'], keep=False)]
duplicates
```

```python
duplicates = df_droped[df_droped.duplicated(subset=['latitude', 'longitude', 'name', 'postalCode'], keep=False)]
duplicates
```

```python
 df_droped2 = df_droped.drop_duplicates(subset=['latitude', 'longitude', 'name', 'postalCode'])
 df_droped2
```

```python
df_droped3 = df_droped2.drop(columns=['categories','country'])
df_droped3
```

```python
df_droped3['dateAdded'] = pd.to_datetime(df_droped3['dateAdded'])
df_droped3['dateUpdated'] = pd.to_datetime(df_droped3['dateUpdated'])
df_droped3
```

```python
df_melted = pd.melt(df_droped3, value_vars=['dateAdded', 'dateUpdated'], var_name='Type', value_name='Date')
```

```python
fig = px.histogram(df_melted, x='Date', color='Type', title='Распределение dateAdded и dateUpdated', nbins=750)
fig.show()
```
